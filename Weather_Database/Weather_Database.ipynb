```python
#Import Deps
import pandas as pd
import numpy as np
import requests
from datetime import datetime
```

```python
#Create a new set of 2,000 random latitudes and longitudes
lats = np.random.uniform(-90.000,90.000, size=2000)
lngs = np.random.uniform(-180.000,180.000, size=2000)
combo_lats_lngs = zip(lats, lngs)
```

```python
coordinates = list(combo_lats_lngs)
```

```python
#Test
coordinates[0]
```

```python
from citipy import citipy
```

```python
#Find nearest City
cities = []

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
#test
cities[0:5]
```

```python
from config import weather_api_key
```

```python
#create URL
url1 = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
import time
```

```python
# Create empty list
city_data = []

# Print top of the logging
print("Beginning of data retrieval      ")
print("---------------------------------")

# Create counters (sets of 50)
record_count =1
set_count=1

# Loop through all cities
for i,city in enumerate(cities):
    
    # Group cities in sets of 50
    if (i%50==0 and i>=50):
        set_count+=1
        record_count=1
        time.sleep(60)
    
    # create endpoint URL for each city
    city_url=url1+"&q="+city.replace(" ","+")
    
    # Log the URL, record, and set numbers and city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    
    # Add 1
    record_count += 1
    

# Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_weatherdesc = city_weather["weather"][0]["description"]
        # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Current Description":city_weatherdesc,
                          "Country": city_country,})

# If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")    
```

```python
# Add data into data frame
weather_city_df=pd.DataFrame(city_data)

weather_city_df
```

```python
# Reorder columns
new_column_order = ["City", "Country", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed", "Current Description"]
weather_city_df = weather_city_df[new_column_order]

#test
weather_city_df
```

```python
# Export data to .csv
output_file = "WeatherPy_Database.csv"
weather_city_df.to_csv(output_file, index_label="City ID")
```
